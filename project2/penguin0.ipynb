```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset("penguins")
df=df.dropna()

def per(a):
    return f"{a.sum()/len(df):.4f}"
```

```python
A=df['species']=='Adelie'
B=df['flipper_length_mm']>=200
print(A.sum())
print(B.sum())
print(f'A,B합사건:{(A|B).sum()}')
print(f'A,B 곱사건: {(A&B).sum()}')
print(f'A의 여사건:{(~A).sum()}')
print(f'AB의 차사건:{(A&(~B)).sum()}')
```
