```python
from sklearn.datasets import fetch_openml
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statistics import mode
from scipy import stats
from scipy.stats import variation

X,y= fetch_openml("titanic",version=1,as_frame=True,return_X_y=True,parser='auto')
```

```python
Pclass_counts=X['pclass'].value_counts().sort_index()
plt.figure()
plt.bar(Pclass_counts.index.astype(str),Pclass_counts.values,color='blue')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.title('Pclass count')
plt.tight_layout()
plt.show()
```

```python
sex_counts=X['sex'].value_counts().sort_index()
plt.figure()
plt.bar(sex_counts.index.astype(str),sex_counts.values,color='magenta')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Sex count')
plt.tight_layout()
plt.show()
```

```python
sibsp=X['sibsp'].value_counts().sort_index()
plt.figure()
plt.bar(sibsp.index.astype(str),sibsp.values,color='blue')
plt.xlabel('SibSp')
plt.ylabel('Count')
plt.title('SibSp count')
plt.tight_layout()
plt.show()
```

```python
parch=X['parch'].value_counts().sort_index()
plt.figure()
plt.bar(parch.index.astype(str),parch.values,color='blue')
plt.xlabel('Parch')
plt.ylabel('Count')
plt.title('Parch count')
plt.tight_layout()
plt.show()
```

```python
age=X['age'].dropna()

plt.boxplot(age)

plt.xlabel('Age')
plt.title('Age')
plt.show()
```

```python
fare_data=X['fare'].dropna()
plt.figure()
plt.hist(fare_data,bins=30,color='cyan')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Fare')
plt.tight_layout()
plt.show()
```

```python
Embarked_counts=X['embarked'].value_counts()
plt.figure()
plt.pie(Embarked_counts,labels=Embarked_counts.index,autopct='%1.1f%%',startangle=90,colors=['blue','green','red'],explode=(0.02,0,0.03))
plt.title('Embarked')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
survived=y.value_counts()
plt.figure()
plt.pie(survived,labels=survived.index,autopct='%1.1f%%',startangle=90,colors=['blue','green'],explode=(0.02,0))
plt.title('Survival Status')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
