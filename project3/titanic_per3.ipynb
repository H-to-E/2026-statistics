```python
from sklearn.datasets import fetch_openml
import pandas as pd
import numpy as np

X,y=fetch_openml("titanic",version=1,as_frame=True,return_X_y=True,parser='auto')

df=pd.concat([X,y],axis=1)
df=df[['pclass','sex','age','sibsp','parch','fare','embarked','survived']]

df['survived']=df['survived'].astype(int)
df['pcalss']=df['pclass'].astype(int)
df= df.dropna().reset_index(drop=True)
```

```python
under=df[df['fare']<50]
middle=df[(df['fare']>50) & (df['fare']<100)]
high=df[df['fare']>=100]

pu=len(under)/len(df)
pm=len(middle)/len(df)
ph=len(high)/len(df)

pus=(under['survived'].sum()/len(under))
pms=(middle['survived'].sum()/len(middle))
phs=(high['survived'].sum()/len(high))

ps=pus*pu+pms*pm+phs*ph

print(f'p(생존|저가)*P(저가)=',(pus*pu).round(4))
print(f'p(생존|중간가)*P(중간가)=',(pms*pm).round(4))
print(f'p(생존|고가)*P(고가)=',(phs*ph).round(4))
print(f'P(생존)=',ps.round(4))
print(f'P(생존)=',f'{1-(df["survived"]==0).sum()/len(df):.4f}')
```
