```python
from sklearn.datasets import fetch_openml
import pandas as pd
import numpy as np

X,y=fetch_openml("titanic",version=1,as_frame=True,return_X_y=True,parser='auto')

df=pd.concat([X,y],axis=1)
df=df[['pclass','sex','age','sibsp','parch','fare','embarked','survived']]

df['survived']=df['survived'].astype(int)
df['pcalss']=df['pclass'].astype(int)
df= df.dropna().reset_index(drop=True)
```

```python
underage=(df['age']<18).sum()/len(df)
undersurvived=((df['age']<18) & (df['survived']==1)).sum()/len(df)
psg=undersurvived/underage

print(f'P(미성년자)',underage.round(4))
print(f'P(생존자|미성년자)',undersurvived.round(4))
print(f'P(생존자|미성년자)/P(미성년자)',psg.round(4)*100)
```
