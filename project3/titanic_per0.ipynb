```python
from sklearn.datasets import fetch_openml
import pandas as pd
import numpy as np

X,y=fetch_openml("titanic",version=1,as_frame=True,return_X_y=True,parser='auto')

df=pd.concat([X,y],axis=1)
df=df[['pclass','sex','age','sibsp','parch','fare','embarked','survived']]

df['survived']=df['survived'].astype(int)
df['pcalss']=df['pclass'].astype(int)
df= df.dropna().reset_index(drop=True)
```

```python
female=(df['sex']=='female').sum()/len(df)
undersurvived=((df['sex']=='female') & (df['survived']==1)).sum()/len(df)
psg=undersurvived/female

print(f'P(여성)',female.round(4))
print(f'P(생존자|여성)',undersurvived.round(4))
print(f'P(생존자|여성)/P(여성)',psg.round(4)*100)
```

```python
female=(df['sex']=='female')
survive=(df['survived']==1)

pera=female.sum()/len(df)
pers=survive.sum()/len(df)
peras=(female&survive).sum()/len(df)

if peras==(pera*pers):
    print('두 사건은 독립')
else:
    print('두 사건은 종속')
```

```python
male=df[df['sex']=='male']
female=df[df['sex']=='female']

pm=len(male)/len(df)
pf=len(female)/len(df)

pms=(male['survived'].sum()/len(male))
pfs=(female['survived'].sum()/len(female))


ps=pms*pm+pfs*pf

print(f'p(생존|남성)*P(남성)=',(pms*pm).round(4))
print(f'p(생존|여성)*P(여성)=',(pfs*pf).round(4))
print(f'P(생존)=',ps.round(4))
print(f'P(생존)=',f'{1-(df["survived"]==0).sum()/len(df):.4f}')
```

```python
groups=sorted(df['pclass'].unique())
pe=(df['survived']==1).sum()/len(df)

for g in groups:
    grp= df[df['pclass']==g]
    pbi=len(grp)/len(df)
    pabi=(grp['survived']==1).sum()/len(grp)
    pbia=(pabi*pbi)/pe

    print(f'pclass가 {g}일 때')
    print(f'사전확률 {pbi:.4f}',)
    print(f'사후확률',pbia.round(4))
```
